## ML Training

To train the ML model, we used the file data.csv (in this same folder) and uploaded it to Google Colab. We then ran this Google Colab Notebook: https://colab.research.google.com/drive/1nhAknWM9pID5gHnqR2hKVyxy07JECJIK?usp=sharing

Although it is possible to download the Google Colab as an ipynb file, the training process is very slow locally because it can only use CPU, not GPU. Google Colab connects automatically to GPU, making the training time less than 10 minutes. Also, it is much easier to use the libraries needed (transformers, torch) in Google Colab than to download them locally. This is why we have not copied over the actual training file, and instead provided the link to the one we used.
